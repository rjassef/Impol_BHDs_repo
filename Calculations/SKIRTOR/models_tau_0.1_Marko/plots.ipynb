## Run through inclinations.

```python
def read_models(tor_oa, con_oa, i, folder="."):

    fname1 = "{}/bHDPol_mrn77_".format(folder)
    fname2 = "tor_oa{:.1f}_con_oa{:.1f}-tauV0.1_i{:.1f}_sed.dat".format(tor_oa, con_oa, i)

    model_both = np.loadtxt(fname1+fname2)
    model_top  = np.loadtxt(fname1+"TopConeOnly_"+fname2)
    model_bottom = np.loadtxt(fname1+"BottomConeOnly_"+fname2)

    return [model_both, model_top, model_bottom]
```

```python
#tor_oa = 60.0
#tor_oa = 55.0
tor_oa = 45.0
#con_oa = 55.0
inc = 90.0

con_oas = np.arange(10.0, tor_oa+0.1-5.0, 5.0)

for con_oa in con_oas:

    models = read_models(tor_oa, con_oa, inc)

    p = []
    for i in range(len(models)):
        p.append((models[i][:,8]**2+models[i][:,9]**2)**0.5/models[i][:,1] * 100)

    fig, ax = plt.subplots(1)

    ax.plot(models[0][:,0], p[0], label="Both Cones")
    ax.plot(models[1][:,0], p[1], label="Top Cone")
    ax.plot(models[2][:,0], p[2], label="Bottom Cone")
    ax.set_title("Cone opening angle {}".format(con_oa))
    ax.set_xlabel("Wavelength (microns)")
    ax.set_ylabel("P (%)")
    ax.legend()
    plt.show()
```
