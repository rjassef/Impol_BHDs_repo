```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_p(cone_type, tor_oa="60.0", cone_oa="55.0"):

    fname = "bHDPol_mrn77_"
    folder = "FullCone/"
    if cone_type=="top":
        fname += "TopConeOnly_"
        folder = "TopConeOnly/"
    elif cone_type=="bottom":
        fname += "BottomConeOnly_"
        folder = "BottomConeOnly/"
    fname += "tor_oa{}_con_oa{}-tauV0.1_i90.0_sed.dat".format(tor_oa, cone_oa)
    
    jlam = 0
    jI = 1
    jQ = 8
    jU = 9

    data = np.loadtxt("{}/{}".format(folder,fname))
    lam = data[:,jlam]
    I   = data[:,jI]
    Q   = data[:,jQ]
    U   = data[:,jU]
    p = (Q**2 + U**2)**0.5 / I * 100.

    return p, lam
    
```

```python
for cone_type in ["full", "top", "bottom"]:
    p, lam = get_p(cone_type)
    plt.plot(lam, p, label=cone_type)
plt.legend()
plt.show()
```

```python
for cone_type in ["full", "top", "bottom"]:
    p, lam = get_p(cone_type, tor_oa="55.0", cone_oa="35.0")
    plt.plot(lam, p, label=cone_type)
plt.legend()
plt.show()
```
