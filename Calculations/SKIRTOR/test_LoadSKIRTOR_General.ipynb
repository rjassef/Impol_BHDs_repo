```python
sk1 = LoadSKIRTOR_General(d03_mod="MW")
```

```python
sk2 = LoadSKIRTOR_General(gs_min=0.005, gs_max=0.25, a=3.5)
print(sk2.root_fname)
```

```python
sk3 = LoadSKIRTOR_General(gs_min=0.005, gs_max=0.25, a=3.5, sil=True)
print(sk3.root_fname)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u 

lam_grid = np.arange(1200., 2800., 20.)*u.AA
iang = 60*u.deg

for sk in [sk1, sk2, sk3]:
    plt.plot(lam_grid, 100*sk.p((iang*np.ones(len(lam_grid)), lam_grid)), label=sk.model_name)

plt.ylabel(r"$P~(\%)$")
plt.xlabel(r"Rest-frame Wavelength $(\rm \AA)$")

plt.legend()

plt.show()
```
