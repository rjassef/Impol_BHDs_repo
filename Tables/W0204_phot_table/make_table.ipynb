```python
import numpy as np
from astropy.table import Table
```

```python
#Read the band names. 
bm_file = Table.read("bandmag.dat", format='ascii')
band_names = bm_file['col1']
```

```python
#Now, read the photometry. 
phot = Table.read("phot.dat", format='ascii')
k = np.argwhere(phot['col1']=='W0204-0506')
print(phot[k])
```

```python
#Finally, we need to read which data was actually used. 
d21 = Table.read("double.21", format='ascii')
d21 = d21[d21['col1']==k[0][0]+1]
d21.show_in_notebook()
```

```python
for i, band_name in enumerate(band_names):
    if d21['col5'][i]==1:
        print("{:25s} & {:5.2f} & {:01.2f} & \\\\".format(band_name, phot['col{}'.format(3+2*i)][k[0][0]], phot['col{}'.format(3+2*i+1)][k[0][0]]))
        #print("{:25s} & {:.2f} \$\pm\$ {:.2f} & \\\\\n".format(band_name, phot['col{}'.format(2+2*i)], phot['col{}'.format(2+2*i+1)]))
```

```python
#Put the photometry in fluxes.
for i, band_name in enumerate(band_names):
    if d21['col5'][i]==1:
        flux = d21['col2'][i] * 1e6
        eflux = d21['col4'][i] * 1e6
        print("{:25s} & {:10.1f} & {:10.1f} & \\\\".format(band_name, flux, eflux))
```
