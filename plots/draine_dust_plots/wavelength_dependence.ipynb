```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u

import os
gd_folder = os.getcwd()+"/../../Calculations/Gas_and_dust/"

import sys 
sys.path.append(gd_folder)
from polWaveDust import PolWaveDust
```

```python
dust_type = "SMC"
pp = PolWaveDust(dust_type, interp_method='cubic')
theta = 70*u.deg
psi = 45*u.deg
lam = np.arange(1000., 3000., 20.)*u.AA
plt.plot(lam, pp.p((lam, theta*np.ones(len(lam)), psi*np.ones(len(lam)))))
```
