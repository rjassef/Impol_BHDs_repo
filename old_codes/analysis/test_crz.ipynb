```python
from astropy.io import fits
from astroscrappy import detect_cosmics
import matplotlib.pyplot as plt
from astropy.visualization import ImageNormalize, LinearStretch, ZScaleInterval
```

```python
#h = fits.open("../proc/reduced_images/science_reduced_img.W0220+0137.60207.R_SPECIAL.chip1.1.fits")
#h = fits.open("../proc/reduced_images/science_reduced_img.W0116-0505.60146.I_BESS.chip1.12.fits")
h = fits.open("../proc/reduced_images/science_reduced_img.W0204-0506.60209.R_SPECIAL.chip1.1.fits")
```

```python
#mask = fits.open("masks/science_reduced_img.W0220+0137.60207.R_SPECIAL.chip1.1.mask.fits")
#mask = fits.open("masks/science_reduced_img.W0116-0505.60146.I_BESS.chip1.12.mask.fits")
mask = fits.open("masks/science_reduced_img.W0204-0506.60209.R_SPECIAL.chip1.1.mask.fits")
```

```python
crmask, clean_im = detect_cosmics(h[0].data, inmask=mask[0].data, objlim=10)
```

```python
norm = ImageNormalize(clean_im[50:190,950:1090], stretch=LinearStretch(), interval=ZScaleInterval())

fig, axs = plt.subplots(1,2, figsize=(10,8))
axs[0].imshow(clean_im[50:190,950:1090], norm=norm, cmap='gray')
axs[1].imshow(h[0].data[50:190,950:1090], norm=norm, cmap='gray')
plt.show()
```
